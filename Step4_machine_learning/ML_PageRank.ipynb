```python
data = final_df[0].to_list()
quater = np.percentile(data, [75, 50 ,25])
```

```python
def convert_to_rating(value):
    if value >= quater[0]:
        return "A"
    elif quater[0] >= value > quater[1]:
        return "B"
    elif quater[1] >= value > quater[2]:
        return "C"
    elif quater[2] > value:
        return "D"
    else:
        return "unknown"
```

```python
list_of_rating = []
for x in data:
    list_of_rating.append(convert_to_rating(x))
final_df['rating'] = list_of_rating
```
