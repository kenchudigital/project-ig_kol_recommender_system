```python
import pandas as pd
```

```python
df = pd.read_csv('kol_info.csv', index_col=[0])
```

```python
df_category = pd.read_csv('result-category.csv', index_col=[0]).reset_index()
```

```python
df_ranking = pd.read_csv('result-ranking.csv', index_col=[0]).reset_index().rename(columns={"index": "ig_id"})
```

```python
df_suggest = pd.read_csv('result-suggest.csv', index_col=[0]).rename(columns={"from": "ig_id"})
```

```python
df1 = df[['ig_id']]
```

```python
df2 = df1.merge(df_category, how='left', on='ig_id')
```

```python
df3 = df2.merge(df_ranking, how='left', on='ig_id')
```

```python
df4 = df3.merge(df_suggest, how='left', on='ig_id')
```
